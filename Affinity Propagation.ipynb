```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
X=dataset.iloc[:,3:5].values
```

```python
X
```

```python
from sklearn.cluster import AffinityPropagation
X,mb= make_blobs(n_samples = 100, centers = [(11,11), (20,20)], n_features =50 , center_box=(1, 2),random_state = 45)
cluster = AffinityPropagation(max_iter=200)
cluster.fit(X)
cluster_centers_ = cluster.cluster_centers_
n_clusters_ = len(cluster_centers_)
y_pred=cluster.predict(X)
colors = plt.cm.viridis(np.linspace(0, 1, n_clusters_))
colors = y_pred
plt.scatter(X[:, 0], X[:, 1], c=y_pred, marker="o", picker=True)
plt.title(f'Estimated number of clusters = {n_clusters_}')
plt.xlabel('clusters')
plt.ylabel('distance')
colors = plt.cm.viridis(np.linspace(0, 1, n_clusters_))
for k, col in zip(range(n_clusters_), colors):
    class_members = y_pred == k
    cluster_center = cluster_centers_[k]
    plt.scatter(
        X[class_members, 0], X[class_members, 1], color=col, marker="."
    )
    plt.scatter(
        cluster_center[0], cluster_center[1], s=14, color=col, marker="o"
    )
    for x in X[class_members]:
        plt.plot(
            [cluster_center[0], x[0]], [cluster_center[1], x[1]], color=col
        )

plt.show()
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import AffinityPropagation
X, _ = make_blobs(n_samples=100, centers=3, n_features=2, random_state=0)
cluster = AffinityPropagation()
cluster.fit(X)
y_pred = cluster.labels_
```
